```python
# making the instance of an object

class PartyAnimal:
    
    def __init__(self):
        self.x = 0
        
    def party(self):
        self.x = self.x + 1
        print('so far x is:',self.x)
        
     # object deleted   
    def __del__(self):
        print('am destructed', self.x)
        
an = PartyAnimal()

# PartyAnimal().party() # is the same as below

an.party()
an.party()
an.party()

an = 40 # object destruction
print('the new value is:',an)

#check what type is an instance
type(an)
dir(an) #what can be done
```

```python
import sqlite3  # Import the SQLite module to work with databases in Python

# Establish a connection to the database (creates 'emaildb.sqlite' if it doesn't exist)
conn = sqlite3.connect('emaildb.sqlite')

# Create a cursor object which allows us to execute SQL commands on the database
cur = conn.cursor()

# Drop the table 'Counts' if it already exists to avoid errors when recreating it
cur.execute('DROP TABLE IF EXISTS Counts')

# Create a new table 'Counts' with two columns: 'email' (text) and 'count' (integer)
cur.execute('CREATE TABLE Counts (org TEXT, count INTEGER)')

# Prompt the user for the file name
fname = input('Enter file name: ')
# Set default file name if none provided
if len(fname) < 1: 
    fname = 'mbox.txt'

# Try opening the file with error handling
try:
    fh = open(fname)
except FileNotFoundError:
    print(f"Error: File '{fname}' not found.")
    exit()

# Loop through each line in the file
for line in fh:
    # If the line does not start with 'From: ', skip it
    if not line.startswith('From: '): 
        continue 

    # Split the line into parts (splits at whitespace by default)
    pieces = line.split()

    # Extract the email address (which is the second part of the line)
    email = pieces[1]
    
    
    # Extract the domain (everything after the '@' symbol)
    org = email.split('@')[1]

    # Select the 'count' value for the domain if it exists in the database
    cur.execute('SELECT count FROM Counts WHERE org = ?', (org, ))
    row = cur.fetchone()  # Fetch one row from the results

    # If the email is not found in the database, insert it with an initial count of 1
    if row is None:
        cur.execute('''INSERT INTO Counts (org, count) VALUES (?, 1)''', (org, ))
    # If the email is found, update its count by adding 1
    else:
        cur.execute('UPDATE Counts SET count = count + 1 WHERE org = ?', (org, ))

    # Commit (save) the changes to the database
    conn.commit()

# Query the database to select the top 10 emails with the highest counts
sqlstr = 'SELECT org, count FROM Counts ORDER BY count DESC LIMIT 10'

# Loop through the rows returned by the SQL query and print each email and its count
for row in cur.execute(sqlstr):
    print(str(row[0]), row[1])

# Close the database connection after the script finishes
conn.close()
```

```python
import sqlite3

# Add a timeout to handle locking issues
conn = sqlite3.connect('trackdb.sqlite', timeout=50)
cur = conn.cursor()

# Make some fresh tables using executescript()
cur.executescript('''
            DROP TABLE IF EXISTS Artist;
            DROP TABLE IF EXISTS Album;
            DROP TABLE IF EXISTS Track;
            DROP TABLE IF EXISTS Genre;

            CREATE TABLE Artist (
                id  INTEGER PRIMARY KEY,
                name    TEXT UNIQUE
            );

            CREATE TABLE Album (
                id  INTEGER PRIMARY KEY,
                artist_id  INTEGER,
                title   TEXT UNIQUE
            );
            
             CREATE TABLE Genre (
                id  INTEGER PRIMARY KEY,
                genre_id  INTEGER,
                title   TEXT UNIQUE
                
            );

            CREATE TABLE Track (
                id  INTEGER PRIMARY KEY,
                title TEXT  UNIQUE,
                genre_id INTEGER,
                album_id  INTEGER,
                len INTEGER, rating INTEGER, count INTEGER
            );
            ''')

# Prompt the user for the file name
ftracks = input('Enter file name: ') #use [tracks.csv] 
if len(ftracks) < 1:
    ftracks = 'tracks.csv'

# Try opening the file with error handling
try:
    ftracks = open(ftracks,  encoding='ISO-8859-1')
except FileNotFoundError:
    print(f"Error: File '{ftracks}' not found.")
    exit()

# Process each line from the file
for line in ftracks:
    line = line.strip()
    pieces = line.split(',')
    if len(pieces) < 6: 
        continue

    name = pieces[0]
    artist = pieces[1]
    album = pieces[2]
    count = pieces[3]
    rating = pieces[4]
    length = pieces[5]
    genre = pieces[6] if len(pieces) > 6 else None  # Handle missing genre

    # print(name, artist, album, count, rating, length, genre)

    # Insert or ignore Artist
    cur.execute('''INSERT OR IGNORE INTO Artist (name) 
        VALUES ( ? )''', (artist,))
    cur.execute('SELECT id FROM Artist WHERE name = ? ', (artist,))
    artist_id = cur.fetchone()[0]

    # Insert or ignore Album
    cur.execute('''INSERT OR IGNORE INTO Album (title, artist_id) 
        VALUES ( ?, ? )''', (album, artist_id))
    
    cur.execute('SELECT id FROM Album WHERE title = ? ', (album,))
    album_id = cur.fetchone()[0]
    
    cur.execute('SELECT id FROM Genre WHERE title = ? ', (genre,))
    row = cur.fetchone()
    
    #conditional check if genre exist
    if row is None:
        # Insert the genre if it doesn't exist
        cur.execute('''INSERT INTO Genre (title) VALUES (?)''', (genre,))
        conn.commit()  # Commit the new genre
        # Fetch the newly inserted genre
        cur.execute('SELECT id FROM Genre WHERE title = ? ', (genre,))
        genre_id = cur.fetchone()[0]
    else:
        genre_id = row[0]

    # Insert or replace Track
    cur.execute('''INSERT OR REPLACE INTO Track
        (title, album_id, len, rating, count, genre_id) 
        VALUES ( ?, ?, ?, ?, ?, ? )''', 
        (name, album_id, length, rating, count, genre_id))

# Commit the changes once, after all rows have been processed
conn.commit()

# Close the database connection after the script finishes
conn.close()
```
